```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (15,5)
data.groupby('State')['Sales'].sum().plot.bar()
```

```python
#1) States generating highest revenues.
state = data.groupby("State")["Sales"].sum()
state = pd.DataFrame(state)
state = state.sort_values("Sales",ascending=0)
state
#We can see Victoria state had most sales in the quarter.
```

```python
business = data.groupby(['State','Time'])['Sales'].sum()
business = pd.DataFrame(business)
business = business.sort_values('Sales',ascending=0)
business
#Analysing which state had more sale in which time period of day.
```

```python
sales_category = data.groupby(['State','Group'])['Sales'].sum()
sales_category = pd.DataFrame(sales_category)
sales_category = sales_category.sort_values('Sales',ascending=0)
sales_category
#Analysing which state had most sale by which group of people.
```

```python
#From the above obtained data, we can see that top 3 states --> Vic, NSW and SA have women playing major role in the sales.
```

```python
units_sold = data.groupby(['State','Group'])['Unit'].sum()
units_sold = pd.DataFrame(units_sold)
units_sold = units_sold.sort_values('Unit',ascending=0)
units_sold
```
